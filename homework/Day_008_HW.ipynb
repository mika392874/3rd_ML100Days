## [ 筆記 ] 什麼是"%matplotlib inline"?
#### "%"為magic function，Python提供了許多魔法命令，使得在IPython環境中操作更加得心應手
-   magic function皆以"%"或者"%%"開頭
-    "%"開頭的為「行命令」-只對命令所在的行有效
-    "%%"開頭的稱為「單元命令」-必須出現在單元的第一行，對整個單元的code進行處理


#### 使用%matplotlib命令-"inline" 
- 使用'%matplotlib inline'命令的功能是可以内嵌繪圖，可省略plt.show()這一步
- 它有一個參數指定圖表的顯示方式

```python
#================================ 以下先不動 ============================================#
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dir_data = '/Users/mikali/Documents/GitHub/mydata/Part01A'
f_app_train = os.path.join(dir_data,'application_train.csv')
app_train = pd.read_csv(f_app_train)
#print(app_train)


#總收入平均值
mean_income = app_train['AMT_INCOME_TOTAL'].mean()
    # mean_income = np.mean(app_train['AMT_INCOME_TOTAL'])
print('總收入平均值：','%.2f'%mean_income)

#總收入標準差
std_income = app_train['AMT_INCOME_TOTAL'].std()
    # std_income = np.std(app_train['AMT_INCOME_TOTAL'])
print('總收入標準差：','%.2f'%std_income)


#總收入中位數
med_income = app_train['AMT_INCOME_TOTAL'].median()
print('總收入中位數：','%.2f'%med_income)
```

```python
# 長條圖-AMT_INCOME_TOTAL：
app_train.loc[app_train['AMT_INCOME_TOTAL']<app_train['AMT_INCOME_TOTAL'].quantile(0.99)]['AMT_INCOME_TOTAL'].hist()

```

## [ Note ]
### app_train.loc[app_train['AMT_INCOME_TOTAL']<app_train['AMT_INCOME_TOTAL'].quantile(0.99)]['AMT_INCOME_TOTAL'].hist()

### 解釋：
##### app_train['AMT_INCOME_TOTAL'] < app_train['AMT_INCOME_TOTAL'].quantile(0.99)
- => 這一段在比較 AMT_INCOME_TOTAL 欄位是否小於 99 分位數，會是一個 Ture/False 的 Series

##### app_train.loc[ ... ]
- => 這一段把上面那個 Ture/False 的 Series 當成條件，會選出 AMT_INCOME_TOTAL 欄位是否小於 99 分位數 的資料

##### app_train.loc[ ... ]['AMT_INCOME_TOTAL']
- => 從符合條件的資料中，取出 AMT_INCOME_TOTAL 欄位

##### app_train.loc[ ... ]['AMT_INCOME_TOTAL'].hist()
- => 畫成長條圖
